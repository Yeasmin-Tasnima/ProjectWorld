```python
import pandas as pd
import numpy as np
```

```python
df_train_ft = pd.read_csv('data/train_features.csv')
df_train_targets = pd.read_csv('data/train_targets_scored.csv')
```

```python
new_df = df_train_targets.set_index('sig_id')
identical = new_df[new_df.duplicated(keep=False)]
identical.shape
```

```python
identical_id = identical.groupby(identical.columns.tolist()).apply(lambda x: tuple(x.index)).tolist()
print('Total set of identical_id: {}'.format(len(identical_id)))
len_ids = [len(i) for i in identical_id]
print('\nlenth of each subsets: \n{}'.format(len_ids))
print('\nidentical_id[5]: {}'.format(identical_id[5]))
```

```python
identical_df = (identical.groupby(identical.columns.tolist()).
                apply(lambda x: tuple(x.index)).reset_index(name='set_id'))
identical_df['length_of_set'] = len_ids
print('identical_df shape: {}'.format(identical_df.shape))
identical_df.head()
```
